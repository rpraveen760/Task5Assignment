```python
#reading the csv files
# load as Pandas dataframe
data1 = pd.read_csv('takehome_users.csv',encoding="ISO-8859-1",parse_dates = ['creation_time'])
data2 = pd.read_csv('takehome_user_engagement.csv',encoding="ISO-8859-1", parse_dates = ['time_stamp'])
```

```python
data1['last_session_creation_time'] = pd.to_datetime(data1['last_session_creation_time'] ,unit='s')
```

```python
data2['date'] = pd.to_datetime(data2.time_stamp.dt.date)# converting time stamp to seconds 
```

```python
def rolling_count(df_group, frequency):
    return df_group.rolling(frequency, on='date')['user_id'].count()
```

```python
data2['visits_7_days'] = data2.groupby('user_id', as_index=False, group_keys=False).apply(rolling_count, '7D')
```

```python
data2
```

```python
data2[data2.visits_7_days >= 3.0]
```

```python
# creating column and label the adopted user based on the condition (any user who visits more than 3 days as adopted user)

# 1 - adopted user 
# 0 - not adopted user

adopted_user = data2.groupby('user_id')['visits_7_days'].max().reset_index()
adopted_user['adopted_user'] = adopted_user['visits_7_days'].apply(lambda x: 1 if x>=3 else 0)
```

```python
adopted_user
```

```python
print(adopted_user['adopted_user'].value_counts())
```

```python
adopted_user.drop('visits_7_days', axis = 1, inplace = True)
adopted_user.rename(columns={"user_id": "object_id"}, inplace=True)
```

```python
adopted_user.set_index("object_id", inplace = True)
```

```python
# joining the adopter user df with the main df 

main_df = data1.join(adopted_user, on = 'object_id', how='left') 
```

```python
main_df
```

```python
main_df.drop(['object_id', 'name', 'email', 'email'], axis = 1, inplace = True) #dropping unwanted columns
```

```python
main_df
```

```python
#one hot encoding
one_hot_encoded_data = pd.get_dummies(main_df, columns = ['creation_source'])
print(one_hot_encoded_data)

main_df = one_hot_encoded_data


```

```python
main_df # after one-hot encoding
```

```python
main_df.drop(['last_session_creation_time', 'creation_time'], axis = 1, inplace = True) #dropping unwanted columns
```

```python
main_df.info()
```

```python
# NA values filled with 0 
main_df['adopted_user'].fillna(0, inplace = True)
main_df['invited_by_user_id'].fillna(0,inplace = True)
```

```python
main_df.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
#creating X and Y variables
y = main_df['adopted_user']
X = main_df.drop(['adopted_user'], axis=1) 
```

```python
X
```

```python
# creating training and testing set

from sklearn.model_selection import train_test_split

# using the train test split function
X_train, X_test, y_train, y_test = train_test_split(X,y ,
								random_state=104,
								test_size=0.2,
								shuffle=True)
```

```python
# # feature scaling
# from sklearn.preprocessing import StandardScaler
# sc_x = StandardScaler()
# xtrain = sc_x.fit_transform(X_train) 
# xtest = sc_x.transform(X_test)
  
# print (xtrain[0:10, :])
```

```python
from sklearn.ensemble import RandomForestClassifier

# creating a RF classifier
rf = RandomForestClassifier(n_estimators = 100)

# Training the model on the training dataset
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
y_pred_train = rf.predict(X_train)

# metrics are used to find accuracy or error
from sklearn import metrics
print()

# using metrics module for accuracy calculation
print("TRAIN ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_train, y_pred_train))
print("TEST ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))

#classification report
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import plot_confusion_matrix
print(classification_report(y_test,y_pred))
cf_matrix = confusion_matrix( y_test,y_pred)

```

```python
#plotting the confusion matrix
import seaborn as sns
import matplotlib.pyplot as plt
ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues',fmt="d")
ax.set_title(' Confusion Matrix \n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');
## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['NOT - Adopted User','Adopted User'])
ax.yaxis.set_ticklabels(['NOT - Adopted User','Adopted User'])
## Display the visualization of the Confusion Matrix.
plt.show()
```
